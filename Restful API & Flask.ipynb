```python
# 1. How do you create a basic Flask application?
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, Flask!'

if __name__ == '__main__':
    app.run()
```

```python
# 2. How do you serve static files like images or CSS in Flask?
# Place your static files in a folder named "static"
# Example: /static/style.css
# In HTML template:
# <link rel="stylesheet" href="{{ url_for('static', filename='style.css') }}">
```

```python
# 3. How do you define different routes with different HTTP methods in Flask?
from flask import request

@app.route('/data', methods=['GET', 'POST'])
def handle_data():
    if request.method == 'POST':
        return 'POST request received'
    return 'GET request received'
```

```python
# 4. How do you render HTML templates in Flask?
from flask import render_template

@app.route('/hello')
def hello():
    return render_template('hello.html')  # Template inside "templates/" folder
```

```python
# 5. How can you generate URLs for routes in Flask using url_for?
from flask import url_for

@app.route('/about')
def about():
    return f"Go to {url_for('home')}"
```

```python
# 6. How do you handle forms in Flask?
from flask import request

@app.route('/form', methods=['GET', 'POST'])
def form():
    if request.method == 'POST':
        name = request.form['name']
        return f"Hello, {name}"
    return '''
        <form method="post">
            <input name="name">
            <input type="submit">
        </form>
    '''
```

```python
# 7. How can you validate form data in Flask?
@app.route('/validate', methods=['POST'])
def validate():
    name = request.form.get('name')
    if not name:
        return 'Name is required'
    return f'Welcome {name}'
```

```python
# 8. How do you manage sessions in Flask?
from flask import session

app.secret_key = 'your_secret_key'

@app.route('/set_session')
def set_session():
    session['user'] = 'Vikram'
    return 'Session set'

@app.route('/get_session')
def get_session():
    return session.get('user', 'Not logged in')
```

```python
# 9. How do you redirect to a different route in Flask?
from flask import redirect

@app.route('/go-home')
def go_home():
    return redirect(url_for('home'))
```

```python
# 10. How do you handle errors in Flask (e.g., 404)?
@app.errorhandler(404)
def page_not_found(e):
    return 'Page not found', 404
```

```python
# 11. How do you structure a Flask app using Blueprints?
# In users.py
from flask import Blueprint

users = Blueprint('users', __name__)

@users.route('/profile')
def profile():
    return 'User Profile'

# In main app
from users import users
app.register_blueprint(users, url_prefix='/user')
```

```python
# 12. How do you define a custom Jinja filter in Flask?
@app.template_filter('reverse')
def reverse_filter(s):
    return s[::-1]

# In template: {{ 'hello' | reverse }}
```

```python
# 13. How can you redirect with query parameters in Flask?
@app.route('/redirect-query')
def redirect_query():
    return redirect(url_for('home', msg='hi'))
```

```python
# 14. How do you return JSON responses in Flask?
from flask import jsonify

@app.route('/api/data')
def api_data():
    return jsonify({'name': 'Vikram', 'status': 'success'})
```

```python
# 15. How do you capture URL parameters in Flask?
@app.route('/user/<username>')
def show_user(username):
    return f'User: {username}'
```
